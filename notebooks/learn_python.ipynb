# Classes
Define a Person class to represent a person with name and age attributes.

```python
class Person:
    def __init__(self, name, age, latitude, longitude):
        self.name = name
        self.age = age
        self.latitude = latitude
        self.longitude = longitude
    
    def happyBirthday(self):
        self.age = self.age + 1

person1 = Person("Alice", 30, 55.6761, 12.5683)
person2 = Person("Bob", 25, 56.1629, 10.2039)
print(person1.name, person1.age, person1.latitude, person1.longitude)
print(person2.name, person2.age, person2.latitude, person2.longitude)
person1.happyBirthday()
print(person1.name, person1.age, person1.latitude, person1.longitude)
```

# Creating Random People
Build a list of 10 Person objects with random names, ages, and home locations near Copenhagen.

```python
people = []
center_lat = 55.6761
center_lon = 12.5683

for _ in range(10):
    person_name = random.choice(name_pool)
    person_age = random.randint(19, 29)
    lat_offset = random.uniform(-0.02, 0.02)
    lon_offset = random.uniform(-0.02, 0.02)
    person_lat = center_lat + lat_offset
    person_lon = center_lon + lon_offset
    people.append(Person(person_name, person_age, person_lat, person_lon))

print(len(people))
print(people[0].name, people[0].age, people[0].latitude, people[0].longitude)
```

# Map of People in Copenhagen
Create a MapLibre map and plot each person's home location.

```python
from anymap_ts import Map

CITY_HALL_LNGLAT = (12.5683, 55.6761)

m = Map(center=CITY_HALL_LNGLAT, zoom=13.5, height="650px")
m.add_basemap("OpenStreetMap.Mapnik")
m.add_3d_buildings(fill_extrusion_color="#888888", fill_extrusion_opacity=0.8)
m.set_sky(sky_color="#00ff84")

for person in people:
    m.add_marker(
        person.longitude,
        person.latitude,
        name=f"person-{person.name}",
        color="#2e7d32",
        popup=f"{person.name}, {person.age}"
    )

m
```
