# Basic Data Types
Create examples and check their types.

```python
count = 3
price = 19.99
name = "Ada"
is_ready = True
print(type(count), type(price), type(name), type(is_ready))
print(name, count, price, is_ready)
```

# Variables and Assignment
Practice naming and reassignment.

```python
student_name = "Lee"
credits = 12
credits = credits + 3
print(student_name, credits)
```

# Arithmetic Operations
Try basic math operators.

```python
a = 10
b = 3
print(a + b, a - b, a * b, a / b, a ** b)
```

# String Manipulation
Concatenate, format, and use common methods.

```python
first = "Hello"
second = "World"
message = first + " " + second
print(message)
print(message.upper())
print(message.lower())
print(message.split())
```

# Lists and Indexing
Create a list, access items, and use methods.

```python
colors = ["red", "green", "blue"]
print(colors[0], colors[-1])
colors.append("yellow")
colors.remove("green")
print(colors)
print(colors[1:3])
```

# Conditionals and Control Flow
Use if, elif, else to make decisions.

```python
temperature = 72
if temperature >= 75:
    print("Warm")
elif temperature >= 60:
    print("Mild")
else:
    print("Cool")
```

# Loops
Use for and while loops.

```python
for i in range(3):
    print("for", i)

count = 0
while count < 3:
    print("while", count)
    count += 1
```

# Functions
Define and call functions.

```python
def greet(name):
    return f"Hello, {name}!"

print(greet("Freshman"))
```
